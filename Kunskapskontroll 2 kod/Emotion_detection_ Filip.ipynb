```python
# Setting picturesize to 48 and naming the correct file path to collect data from
picture_size = 48
folder_path = "./face-expression-recognition-dataset/images/"
```

# Data Exploration and Augmentation

Small exploration of the training data and also of the amount of pictures in each class. As we can see the dataset is not balanced as there for example is much more happy pictures than disgust pictures. Data Augumentation using ImageDataGenerator to create a more varied training data using rotation, zoom, flip and so on. 

```python
# Plotting a few pictures from the training dataset
expression = 'sad'

plt.figure(figsize= (12,12))
for i in range(1, 10, 1):
    plt.subplot(3,3,i)
    img = load_img(folder_path+"train/"+expression+"/"+
                  os.listdir(folder_path + "train/" + expression)[i], target_size=(picture_size, picture_size))
    plt.imshow(img)
plt.show()
```

```python
# Counting number of images in train set for each expression
for expression in os.listdir(folder_path + "train"):
    print(str(len(os.listdir(folder_path + "train/" + expression))) + " " + expression + " images")
```

```python
# Using ImageDataGenerator for data augumentation
batch_size  = 128

datagen_train = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

datagen_val = ImageDataGenerator(rescale=1./255)

train_set = datagen_train.flow_from_directory(folder_path+"train",
                                              target_size = (picture_size,picture_size),
                                              color_mode = "grayscale",
                                              batch_size=batch_size,
                                              class_mode='categorical',
                                              shuffle=True)


test_set = datagen_val.flow_from_directory(folder_path+"validation",
                                              target_size = (picture_size,picture_size),
                                              color_mode = "grayscale",
                                              batch_size=batch_size,
                                              class_mode='categorical',
                                              shuffle=False)
```

# Model Building

Building the sequential model. I choose to use 5 CNN layers, then global average pooling, 2 fully connected layers and finally the ouput layer. I tried different models, more layers, fewer layers and so on. I set myself a goal to get over 50% val accuracy and ended up settling with this model. Its a pretty big model with lots of layers, I also used 4 CNN layers and 1 fully connected but only managed to get around 40% val accuracy.

After creating the model its compiled, then a checkpoint saving the best model with the highest val accuracy is created. Also early stopping and reduce learningrate callbacks are created monitoring the val_loss. Before fitting the model the class weights are calculated and saved in a dictionary. Like we saw from the data exploration the classes had different amount of data per class.

```python
# Building model
no_of_classes = 7

model = Sequential()
model.add(Input(shape=(48, 48, 1)))

# 1st CNN layer
model.add(Conv2D(64, (3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# 2nd CNN layer
model.add(Conv2D(128, (3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# 3rd CNN layer
model.add(Conv2D(256, (3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# 4th CNN layer
model.add(Conv2D(512, (3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# 5th CNN layer
model.add(Conv2D(1024, (3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# Global Average Pooling
model.add(GlobalAveragePooling2D())

# 1st Fully connected layer
model.add(Dense(512))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5))

# 2nd Fully connected layer
model.add(Dense(256))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5))

# Output layer
model.add(Dense(no_of_classes, activation='softmax'))
```

```python
model.summary()
```

```python
# Compiling model
model.compile(loss='categorical_crossentropy', optimizer = Adam(learning_rate=0.001), metrics=['accuracy'])
```

```python
# Creating Checkpoint, Early Stopping and Reduce Learningrate
checkpoint = ModelCheckpoint("./model_checkpoint.h5", monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')

early_stopping = EarlyStopping(monitor='val_loss',
                          min_delta=0,
                          patience=3,
                          verbose=1,
                          restore_best_weights=True
                          )

reduce_learningrate = ReduceLROnPlateau(monitor='val_loss',
                              factor=0.2,
                              patience=3,
                              verbose=1,
                              min_delta=0.0001)

callbacks_list = [early_stopping,checkpoint,reduce_learningrate]

epochs = 50
```

```python
#Importing class weights to deal with the different amount of data in each class.
from sklearn.utils import class_weight

# Compute class weights
class_weights = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(train_set.classes),
    y=train_set.classes
)

# Convert class weights to a dictionary
class_weights = {i : class_weights[i] for i in range(len(class_weights))}
```

# Model fitting and plots

Fitting the model gets us the results below. I wanted a val accuracy over 50% which was achieved. Looking at the plots the val line is a little bit spiky, but it still follows the training curve. I tried different methods to get rid of the spikes like changing learning rate, using more dropout etc. But its always a bit spiky but when testing the model using camera it still works good, so my conclusion is that its not affecting the model too much. If I had more time I would go further into it too see if there is a solution. Also If i had more time I would use one more dataset and train the model on more data to get an even higher val accuracy.

```python
# Fitting model
history = model.fit(
    train_set,
    steps_per_epoch=train_set.n // train_set.batch_size,
    epochs=epochs,
    validation_data=test_set,
    validation_steps=test_set.n // test_set.batch_size,
    class_weight=class_weights,
    callbacks=callbacks_list
)
```

```python
# Plotting Loss and Accuracy
plt.figure(figsize=(20,10))
plt.subplot(1, 2, 1)
plt.suptitle('Optimizer : Adam', fontsize=10)
plt.ylabel('Loss', fontsize=16)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend(loc='upper right')

plt.subplot(1, 2, 2)
plt.ylabel('Accuracy', fontsize=16)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend(loc='lower right')
plt.show()
```
